## **Overview**

*   Matplotlib is a popular data visualization library in Python.
* It provides a wide range of tools for creating static, animated, and interactive plots.




## **Key Features**


*   Matplotlib allows to create customise plots.
*   Supports various plots such as **Line Plot**, **Scatter Plot** , **Histogram** and many more
* Allows Customization of Plots, such as colors, markers, labels, titles, axes, and legends.

Example : Import the Matplotlib library and pyplot module

```python
import matplotlib.pyplot as plt
```

create a simple line plot âž–

```python
x = [1,2,3,4]    # x-axis values

y = [10,20,30,40]   # y-axis values

plt.plot(x ,y)
plt.show()
```

### Customize plot

```python
x = [1,2,3,4]    # x-axis values

y = [10,20,30,40]   # y-axis values

plt.plot(x ,y,label = "Line Plot" , color = "green")
plt.xlabel('X-axis Values')   # X-axis label
plt.ylabel('Y-axis Values')   # Y-axis label
plt.title('Simple Line Plot') # Title of plot
plt.legend()
plt.show()
```

* Matplotlib is a powerful data visualization library that enables you to create a wide variety of plots.
* It offers extensive customization options and is widely used in the scientific and data analysis communities.
